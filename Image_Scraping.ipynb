```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os
```

```python
excel_data = pd.read_excel("Python_Image_Extraction.xlsx", engine = 'openpyxl')

excel_data.set_index('Constituency No.', inplace = True)

for const in range(59,295):

    win_party = excel_data.loc[const, '2021 Winner']
    
    win_candidate = excel_data.loc[const, 'Winning Candidate 2021']

    url1 = f'https://affidavit.eci.gov.in/CandidateCustomFilter?_token=MROovv5tWcQr9GvE3tV3XSMfHRWCyBzXkzin5dCI&electionType=10-AC-GENERAL-3-16&election=10-AC-GENERAL-3-16&states=S25&constId={const}&page=1'

    r1 = requests.get(url1)
    soup1 = BeautifulSoup(r1.text, 'html.parser')

    items = soup1.find_all('li', {'class':'page-item'})
    
    if(len(items)==0):
        No_pages = 1
    else:
        No_pages = len(items)-2
        

    link_images=[]
    paragraphs=[]

    for page in range(1,No_pages+1):
        url2 = f'https://affidavit.eci.gov.in/CandidateCustomFilter?_token=MROovv5tWcQr9GvE3tV3XSMfHRWCyBzXkzin5dCI&electionType=10-AC-GENERAL-3-16&election=10-AC-GENERAL-3-16&states=S25&constId={const}&page={page}'
        r2 = requests.get(url2)
        soup2 = BeautifulSoup(r2.text, 'html.parser')

        table = soup2.find_all('tr')
        #print(table)


        for x in table:
            paragraphs.append(str(x))

    strings_with_substring = [string for string in paragraphs if win_party in string]
    strings_with_substring = str(strings_with_substring)
    soup = BeautifulSoup(strings_with_substring, "html.parser")
    soup = BeautifulSoup(str(set(soup.find_all('img'))), "html.parser")
    images = soup.find_all('img')
    #print(images)

    for image in images:
        link=image['src']
        if(link==''):
            print("Empty")
        with open("D:\Working_Folder\EL_Tableau\Photos\\" + win_candidate + '.jpg','wb') as f:
            im = requests.get(link)
            f.write(im.content)


    print(const)
```
