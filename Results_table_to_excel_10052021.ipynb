```python
results1 = []



## 1 to 295 constituencies (Total constituencies = 294 so looping till 295)

for i in range(1,295):

    url = f'https://results.eci.gov.in/Result2021/RoundwiseS25{i}.htm?ac={i}'
    response = requests.get(url)
    html = response.content

   

    soup = BeautifulSoup(html)
    #   tables = soup.findAll("table")



    results2 = []
    results3 = []
    
    
    
## 1 to 50 rounds of counting in each constituency

    for j in range(1,50):
        results = soup.find(id= f'tab{j}')
        if (results == None):
            break
        else:
            results1 = results.find_all('td')


            for a in results1:
                results3.append(a.text)

                
                
## Code snippet for adding round no. after every N spaces in each candidate details

            test_list = results3
            k = j
            N1 = 5
            results3 = list(chain(*[test_list[u : u+N1] + [i] 
                        if len(test_list[u : u+N1]) == N1 
                        else test_list[u : u+N1] 
                        for u in range(0, len(test_list), N1)]))
            
            test_list = results3
            N2 = 6
            results3 = list(chain(*[test_list[u : u+N2] + [k] 
                        if len(test_list[u : u+N2]) == N2 
                        else test_list[u : u+N2] 
                        for u in range(0, len(test_list), N2)]))

            
## Code snippet to insert round no. at the beginning of each constituency round of counting
## Code snippet to delete the extra round no. added at the end of each round of counting

            results3.insert(0,i)
            results3.insert(1,j)
            results3.pop(-1)
            results3.pop(-1)
            
            
            
## Adding the result of the current round with the total and emptying the results3 for inserting new value for next round

        results2.extend(results3)
        results3 = []


## Splitting the complete result of each constituency into chunks of 6 coz there are 6 columns

    def chunks(lst, n):
        """Yield successive n-sized chunks from lst."""
        for i in range(0, len(lst), n):
            yield lst[i:i + n]


    results2 = list(chunks(results2, 7))

## Defining dataframe and writing it to excel

    df = df.append(pd.DataFrame(results2, columns = columns), ignore_index=True)
    print(i)   
    
```
